## Titanic data

Basic functions alayzing titanic data





```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_excel('./data/titanic3.xls', index_col=None, na_values=['NA'])
titanic_df.head()
```

```python
titanic_df.describe()
```

```python
titanic_df.drop(["ticket", "cabin", "boat", "body"], axis=1).head()
```

```python
pd.value_counts(titanic_df["survived"]).plot.pie()
```

```python
titanic_df["survived"].mean()
```

```python
titanic_df.groupby(["sex"]).mean()
```

```python
titanic_df.groupby(["sex", "pclass"]).mean()
```

```python
titanic_df[titanic_df["age"]<18].groupby(["sex", "pclass"]).mean()
```
